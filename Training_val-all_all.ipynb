```python
train_data = shuffle(data[data.date_block_num <= 33])

frac = round(train_data.shape[0] * 0.9)
X_train = train_data[:frac].drop(['item_cnt_month'], axis=1)
Y_train = train_data[:frac]['item_cnt_month']
X_valid = train_data[frac:].drop(['item_cnt_month'], axis=1)
Y_valid = train_data[frac:]['item_cnt_month']
del train_data

X_test = data[data.date_block_num == 34].drop(['item_cnt_month'], axis=1)
del data
gc.collect();
```

# XGBoost Training

```python
model = XGBRegressor(
    max_depth=8,
    n_estimators=100,
    min_child_weight=300, 
    colsample_bytree=0.8, 
    subsample=0.8, 
    eta=0.3,    
    seed=42,
    n_jobs=8)

model.fit(
    X_train, 
    Y_train, 
    eval_metric="rmse", 
    eval_set=[(X_train, Y_train), (X_valid, Y_valid)], 
    verbose=True, 
    early_stopping_rounds=25)
```

```python
def plot_features(booster, figsize):
    fig, ax = plt.subplots(1,1,figsize=figsize)
    return plot_importance(booster=booster, ax=ax)
plot_features(model, (10,25))
```

# Output File

```python
Y_pred = model.predict(X_valid).clip(0, 20)
Y_test = model.predict(X_test).clip(0, 20)

test  = pd.read_csv('./input/test.csv').set_index('ID')
submission = pd.DataFrame({
    "ID": test.index, 
    "item_cnt_month": Y_test
})
submission.to_csv('submission_val-all_all_es25.csv', index=False)
```
